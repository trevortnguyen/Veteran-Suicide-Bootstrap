# 1. Data Pre-Processing

## 1.1. Load Libraries

```r
library(boot)
library(ggplot2)
library(kableExtra)
library(reshape2)
library(tidyr)
```

## 1.2. Read Files

```r
data2005 <- read.csv("2005.csv")
data2006 <- read.csv("2006.csv")
data2007 <- read.csv("2007.csv")
data2008 <- read.csv("2008.csv")
data2009 <- read.csv("2009.csv")
data2010 <- read.csv("2010.csv")
data2011 <- read.csv("2011.csv")
```

## 1.3. Data Cleaning

```r
# Rename overall_pop_18 column
colnames(data2005)[4] <- "overall_pop"
colnames(data2006)[4] <- "overall_pop"
colnames(data2007)[4] <- "overall_pop"
colnames(data2008)[4] <- "overall_pop"
colnames(data2009)[4] <- "overall_pop"
colnames(data2010)[4] <- "overall_pop"
colnames(data2011)[4] <- "overall_pop"
```

```r
# Create 2 new columns for civilian population and suicides
data2005$civ_pop <- data2005$overall_pop - data2005$vet_pop
data2006$civ_pop <- data2006$overall_pop - data2006$vet_pop
data2007$civ_pop <- data2007$overall_pop - data2007$vet_pop
data2008$civ_pop <- data2008$overall_pop - data2008$vet_pop
data2009$civ_pop <- data2009$overall_pop - data2009$vet_pop
data2010$civ_pop <- data2010$overall_pop - data2010$vet_pop
data2011$civ_pop <- data2011$overall_pop - data2011$vet_pop

data2005$civ_suicides <- data2005$all_suicides - data2005$vet_suicides
data2006$civ_suicides <- data2006$all_suicides - data2006$vet_suicides
data2007$civ_suicides <- data2007$all_suicides - data2007$vet_suicides
data2008$civ_suicides <- data2008$all_suicides - data2008$vet_suicides
data2009$civ_suicides <- data2009$all_suicides - data2009$vet_suicides
data2010$civ_suicides <- data2010$all_suicides - data2010$vet_suicides
data2011$civ_suicides <- data2011$all_suicides - data2011$vet_suicides
```

```r
# Subset columns
data_2005 <- subset(data2005, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))

data_2006 <- subset(data2006, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))

data_2007 <- subset(data2007, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))

data_2008 <- subset(data2008, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))

data_2009 <- subset(data2009, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))

data_2010 <- subset(data2010, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))

data_2011 <- subset(data2011, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, 
                                         civ_suicides, all_suicides, vet_rate, civ_rate))
```

# 3. Bootstrapping 

```r
set.seed(3) # for reproducibility
```

## 3.1. Bootstrapping Mean

```r
mean_function <- function(data, indices) {
  mean <- mean(data[indices])
  return (mean)
}
```

### 3.1.1. Bootstrapping Mean Veteran Suicide Rate

```r
boot_2005_mean_vet <- boot(data = data_2005$vet_rate, statistic = mean_function, R = 1000)
boot_2006_mean_vet <- boot(data = data_2006$vet_rate, statistic = mean_function, R = 1000)
boot_2007_mean_vet <- boot(data = data_2007$vet_rate, statistic = mean_function, R = 1000)
boot_2008_mean_vet <- boot(data = data_2008$vet_rate, statistic = mean_function, R = 1000)
boot_2009_mean_vet <- boot(data = data_2009$vet_rate, statistic = mean_function, R = 1000)
boot_2010_mean_vet <- boot(data = data_2010$vet_rate, statistic = mean_function, R = 1000) 
boot_2011_mean_vet <- boot(data = data_2011$vet_rate, statistic = mean_function, R = 1000)

boot_ci_2005_mean_vet <- boot.ci(boot_2005_mean_vet, type = "perc")
boot_ci_2006_mean_vet <- boot.ci(boot_2006_mean_vet, type = "perc")
boot_ci_2007_mean_vet <- boot.ci(boot_2007_mean_vet, type = "perc")
boot_ci_2008_mean_vet <- boot.ci(boot_2008_mean_vet, type = "perc")
boot_ci_2009_mean_vet <- boot.ci(boot_2009_mean_vet, type = "perc")
boot_ci_2010_mean_vet <- boot.ci(boot_2010_mean_vet, type = "perc")
boot_ci_2011_mean_vet <- boot.ci(boot_2011_mean_vet, type = "perc")
```

### 3.1.2. Bootstrapping Mean Civilian Suicide Rate

```r
boot_2005_mean_civ <- boot(data = data_2005$civ_rate, statistic = mean_function, R = 1000)
boot_2006_mean_civ <- boot(data = data_2006$civ_rate, statistic = mean_function, R = 1000)
boot_2007_mean_civ <- boot(data = data_2007$civ_rate, statistic = mean_function, R = 1000)
boot_2008_mean_civ <- boot(data = data_2008$civ_rate, statistic = mean_function, R = 1000)
boot_2009_mean_civ <- boot(data = data_2009$civ_rate, statistic = mean_function, R = 1000)
boot_2010_mean_civ <- boot(data = data_2010$civ_rate, statistic = mean_function, R = 1000) 
boot_2011_mean_civ <- boot(data = data_2011$civ_rate, statistic = mean_function, R = 1000)

boot_ci_2005_mean_civ <- boot.ci(boot_2005_mean_civ, type = "perc")
boot_ci_2006_mean_civ <- boot.ci(boot_2006_mean_civ, type = "perc")
boot_ci_2007_mean_civ <- boot.ci(boot_2007_mean_civ, type = "perc")
boot_ci_2008_mean_civ <- boot.ci(boot_2008_mean_civ, type = "perc")
boot_ci_2009_mean_civ <- boot.ci(boot_2009_mean_civ, type = "perc")
boot_ci_2010_mean_civ <- boot.ci(boot_2010_mean_civ, type = "perc")
boot_ci_2011_mean_civ <- boot.ci(boot_2011_mean_civ, type = "perc")
```

### 3.1.3. Hypothesis Testing on Mean Veteran and Civilian Suicide Rates

```r
# Parametric
t_test_mean_2005 <- t.test(boot_2005_mean_vet$t, boot_2005_mean_civ$t, alternative = "greater")
t_test_mean_2006 <- t.test(boot_2006_mean_vet$t, boot_2006_mean_civ$t, alternative = "greater")
t_test_mean_2007 <- t.test(boot_2007_mean_vet$t, boot_2007_mean_civ$t, alternative = "greater")
t_test_mean_2008 <- t.test(boot_2008_mean_vet$t, boot_2008_mean_civ$t, alternative = "greater")
t_test_mean_2009 <- t.test(boot_2009_mean_vet$t, boot_2009_mean_civ$t, alternative = "greater")
t_test_mean_2010 <- t.test(boot_2010_mean_vet$t, boot_2010_mean_civ$t, alternative = "greater")
t_test_mean_2011 <- t.test(boot_2011_mean_vet$t, boot_2011_mean_civ$t, alternative = "greater")

# Nonparametric 
wilcox_test_mean_2005 <- wilcox.test(boot_2005_mean_vet$t, boot_2005_mean_civ$t, alternative = "greater")
wilcox_test_mean_2006 <- wilcox.test(boot_2006_mean_vet$t, boot_2006_mean_civ$t, alternative = "greater")
wilcox_test_mean_2007 <- wilcox.test(boot_2007_mean_vet$t, boot_2007_mean_civ$t, alternative = "greater")
wilcox_test_mean_2008 <- wilcox.test(boot_2008_mean_vet$t, boot_2008_mean_civ$t, alternative = "greater")
wilcox_test_mean_2009 <- wilcox.test(boot_2009_mean_vet$t, boot_2009_mean_civ$t, alternative = "greater")
wilcox_test_mean_2010 <- wilcox.test(boot_2010_mean_vet$t, boot_2010_mean_civ$t, alternative = "greater")
wilcox_test_mean_2011 <- wilcox.test(boot_2011_mean_vet$t, boot_2011_mean_civ$t, alternative = "greater")
```

# 6. Binomial and Normal Approximation

```r
data2005$year <- '2005'
data2006$year <- '2006'
data2007$year <- '2007'
data2008$year <- '2008'
data2009$year <- '2009'
data2010$year <- '2010'
data2011$year <- '2011'
```

```r
total_data <- rbind(data2005, data2006, data2007, data2008, data2009, data2010, data2011)

total_data$civ_pop <- total_data$overall_pop - total_data$vet_pop

total_data$civ_suicides <- total_data$all_suicides - total_data$vet_suicides
colnames(total_data)[4] <- "overall_pop"

total_data <- subset(total_data, select = c(state, vet_pop, civ_pop, overall_pop, vet_suicides, civ_suicides, all_suicides, vet_males, vet_females, vet_rate, civ_rate, year))
```

```r
# Data is prepared for analysis
n <- 350

x1 <- sample(1:350, 350, replace = TRUE)

prop_difs <- total_data$vet_rate[x1] - total_data$civ_rate[x1]
#print(prop_difs)

mu_1 <- mean(prop_difs)  
sigma_1 <- sd(prop_difs)

## ^^ These are the bootstrap replicates for the mean and standard deviation of proportion difference in bootstrap sample of n = 350
```

```r
# We may also want to consider assumptions about the distributions of these proportion differences (ie in relation to binomial distribution)
B <- 20
mean_star <- c()

for (b in 1:B) {
  indices <- sample(1:350, 350, replace = TRUE)
  prop_difs <- total_data$vet_rate[indices] - total_data$civ_rate[indices]
  
  mean_star <- append(mean_star, mean(prop_difs))
}


mean_estimate_ssq <- sum((mean_star - mean(mean_star))^2)
se_B_hat <- sqrt((1/(B-1)) * mean_estimate_ssq)


# The variable se_B_hat is the bootstrap estimate of the standard error of the mean of the proportion difference with B = 20. As we can see, it's quite small (< 1) and may be taken to follow the t distribution.  If we consider that the bootstrap estimate of the mean is mean(mean_star) = 16.66:

# test statistic: t0 = (16.66 - 0)/(.5434) = 30.64 on B-1 = 19 degrees of freedom

# The above is (more than) sufficient evidence to assert that veterans have a higher suicide rate than non-veterans.

# I THINK that the SE for bootstrap estimate given in class means that it's t-distributed.
```

```r
# Mean suicides is simply the observed number of suicides
total_data$p_vet <- (1/100000) * total_data$vet_rate
total_data$comp_p_vet <- 1-total_data$p_vet

total_data$p_civ <- (1/100000) * total_data$civ_rate
total_data$comp_p_civ <- 1-total_data$p_civ

total_data$sigma_sq_vet <- total_data$vet_pop * total_data$p_vet * (1-total_data$p_vet)
total_data$sigma_sq_civ <- total_data$civ_pop * total_data$p_civ * (1-total_data$p_civ)

total_data$sigma_sq_p_vet <- total_data$sigma_sq_vet/((total_data$vet_pop)^2)

total_data$sigma_sq_p_civ <- total_data$sigma_sq_civ/((total_data$civ_pop)^2)

total_data$p_dif <- total_data$p_vet - total_data$p_civ
total_data$sigma_sq_dif <- total_data$sigma_sq_p_civ + total_data$sigma_sq_p_vet

## ^^ The mean and variance for normally distributed difference in suicide rates
```

```r
total_data$sigma_sq_vet <- NULL
total_data$sigma_sq_civ <- NULL
total_data$sigma_sq_p_vet <- NULL
total_data$sigma_sq_p_civ <- NULL
total_data$comp_p_vet <- NULL
total_data$comp_p_civ <- NULL
total_data$p_vet <- NULL
total_data$p_civ <- NULL

## ^^ Cleaning up the dataframe; no need for intermediate values anymore
```

```r
B <- 8
minlist <- c()
maxlist <- c()
```

```r
for (b in 1:B) {
  indices <- sample(1:350, 350, replace = TRUE)
  sum_of_point_estimates <- sum(total_data$p_dif[indices])
  big_variance <- sum(total_data$sigma_sq_dif[indices])
  bootstrap_dif_point_est <- sum_of_point_estimates/350
  bootstrap_dif_variance <- big_variance/(350^2)
  bootstrap_dif_sd <- sqrt(bootstrap_dif_variance)
  
  CI_min <- 100000 * (bootstrap_dif_point_est - 1.96*bootstrap_dif_sd)
  CI_max <- 100000 * (bootstrap_dif_point_est + 1.96*bootstrap_dif_sd)
  
  minlist <- append(minlist, CI_min)
  maxlist <- append(maxlist, CI_max)
}
```

```r
for (i in 1:B) {
  cat("Index of Bootstrap Sample:", as.integer(i), '\n')
  cat('[', as.character(minlist[i]), ', ', as.character(maxlist[i]), ']', "\n", sep = '')
}
```
